```python
import requests
import json
import pandas as pd
```

```python
response = requests.get(url= "https://www.datos.gov.co/resource/39pv-4vq3.json")
#se carga la api
```

```python
datos =json.loads(response.text) #nos trae la información como un diccionario de datos
```

```python
df = pd.DataFrame(datos)
```

```python
df2=df.drop_duplicates()
print(df2.shape) # no se encuentran duplicados
```

```python
#se realiza un renombre de las variables mal escritas

df = df.rename(columns={'area_defunci_n': 'area_defunción'})
df = df.rename(columns={'sitio_defunci_n': 'sitio_defunción'})
df = df.rename(columns={'nombre_instituci_n': 'nombre_institución'})
df = df.rename(columns={'a_o': 'año'})
df = df.rename(columns={'hora_defunci_n': 'hora_defunción'})
df = df.rename(columns={'fecha_defunci_n': 'fecha_defunción'})
df = df.rename(columns={'tipo_defunci_n': 'tipo_defunción'})
df = df.rename(columns={'ultimo_a_o_de_formaci_n': 'ultimo_año_de_formación'})
df = df.rename(columns={'ocupaci_n_fallecido': 'ocupación_fallecido'})
df = df.rename(columns={'pa_s_residencia': 'país_residencia'})
df = df.rename(columns={'r_gimen_seguridad': 'régimen_seguridad'})
df = df.rename(columns={'pertenencia_tnica': 'pertenencia_étnica'})
```

```python
df.columns # Se comprueba el cambio de nombre de las variables
```

```python
# Se Convertierte la columna de fecha_defunción al tipo de dato datetime

df['fecha_defunción'] = pd.to_datetime(df['fecha_defunción'])

# Crear nuevas columnas para mes y año
df['mes'] = df['fecha_defunción'].dt.month
df['año'] = df['fecha_defunción'].dt.year

# Muestra las primeras filas del DataFrame resultante
print(df[['fecha_defunción', 'mes', 'año']].head())
```
